# Reading Data

```python
df = pd.read_csv("./datasets/pre_processed.csv")
df.head()
```

# Test n Train Split

```python
y = df["SalePrice"]
X = df.drop("SalePrice", axis=1)

# test train split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
```

# Feature Scalling

```python
#Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Modelling

- Ridge
- Lasso
- Elastic net
- Random forest
- Support vector regressor
- Gradient boosting
- Light GBM
- XGBoost

```python
svr = SVR()
lasso = Lasso(alpha=0.001)
ridge = Ridge(alpha=10)
rf = RandomForestRegressor()
gbr = GradientBoostingRegressor()
lightgbm = LGBMRegressor()
xgboost = XGBRegressor()
```

```python
# Evaluation Metrices

def MAPE(y_true, y_pred):
    """
    Mean Absolute Percentage Error
    """
    
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


def MAE(actual, predicted):
    """
    Mean Absolute Error
    """

    return np.mean(np.abs(actual - predicted))
```

```python
models = [ridge, lasso, rf, svr, gbr, lightgbm, xgboost]
r2s = []
r2_adjs = []
mapes = []
maes = []

for model in models:
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)

    r2=r2_score(y_pred, y_test)
    n =len(X_train)
    r2_adj =1- (1-r2)*(n-1)/(n-(13+1))

    r2s.append(r2)
    r2_adjs.append(r2_adj)
    mapes.append(MAPE(y_pred, y_test))
    maes.append(MAE(y_pred, y_test))



models_df = pd.DataFrame({'Model': ['Ridge', 'Lasso', 'Random Forest', 'SVR', 'Gradient Boosting', 'Light GBM', 'XGBoost'],
                         'R square': r2, 'Adjusted R square': r2_adjs, "MAPE": mapes, "MAE": maes})
models_df.sort_values(by = 'R square', inplace = True, ignore_index = True)
models_df.set_index("Model")

```
